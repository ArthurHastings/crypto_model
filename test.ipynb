```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
df_sentiment = pd.read_csv("filled_headline_sentiments.csv")
df_price = pd.read_csv("apple_price_360d.csv")
```

```python
min_sentiment_date = df_sentiment["Date"].min()

# Filter df_price to start from that date
df_price = df_price[df_price["Date"] >= min_sentiment_date].reset_index(drop=True)

# Display updated df_price
df_price
```

```python
df_price["Sentiment"] = 0
```

```python
def compute_sentiment(date):
    matching_rows = df_sentiment[df_sentiment["Date"] == date]
    
    if not matching_rows.empty:
        dominant_sentiments = matching_rows[["Negative", "Neutral", "Positive"]].idxmax(axis=1)
        
        sentiment_scores = dominant_sentiments.map({"Negative": -1, "Neutral": 0, "Positive": 1})
        
        return sentiment_scores.mean()
    
    return 0.0

df_price["Date"] = pd.to_datetime(df_price["Date"])
df_sentiment["Date"] = pd.to_datetime(df_sentiment["Date"])

df_price["Sentiment"] = df_price["Date"].apply(compute_sentiment)

df_price.head()
```
